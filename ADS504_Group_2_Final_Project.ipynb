```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('/universal_top_spotify_songs.csv', on_bad_lines='skip', engine='python')

# Display the first few rows of the dataset
print(df.head())

# Display basic information about the dataset
print(df.info())

# Display summary statistics of the dataset
print(df.describe())

# Check for missing values
missing_data = df.isnull().sum()
missing_percentage = (df.isnull().sum() / len(df)) * 100
print(missing_data)
print(missing_percentage)

# Visualize the missing data with an enhanced heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.isnull(), cbar=True, cmap='plasma', yticklabels=False, cbar_kws={'label': 'Missing Data'})
plt.title('Missing Data Heatmap', fontsize=20)
plt.xlabel('Features', fontsize=15)
plt.ylabel('Records', fontsize=15)
plt.xticks(rotation=45)
plt.show()

# Handle missing values
# Example: Fill missing values with the mean (numeric columns) or mode (categorical columns)
for column in df.columns:
    if df[column].dtype == 'object':
        df[column].fillna(df[column].mode()[0], inplace=True)
    else:
        df[column].fillna(df[column].mean(), inplace=True)

# Verify missing values have been handled
print(df.isnull().sum())

# Visualize the distribution of numerical features
df.hist(bins=30, figsize=(15, 10))
plt.tight_layout()
plt.show()

# Visualize the distribution of categorical features
categorical_columns = df.select_dtypes(include=['object']).columns
for column in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=column)
    plt.title(f'Distribution of {column}')
    plt.xticks(rotation=45)
    plt.show()

# Visualize correlations using a heatmap
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

# Encode categorical variables (if any)
df_encoded = pd.get_dummies(df, drop_first=True)

# Feature scaling (if necessary)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_encoded)

# Convert the scaled features back to a DataFrame
df_scaled = pd.DataFrame(scaled_features, columns=df_encoded.columns)

# Display the first few rows of the scaled DataFrame
print(df_scaled.head())

# Visualize the heatmap of the entire dataset (scaled)
plt.figure(figsize=(12, 8))
sns.heatmap(df_scaled, cmap='coolwarm', cbar=True)
plt.title('Heatmap of Scaled Data')
plt.show()
```
